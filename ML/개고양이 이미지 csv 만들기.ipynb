```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image

# for문 만들기
# 1. 파일 가져와서 resize
# 2. 흑백 변환 및 2d로 변환
# 3. df에 append 하면서 label 붙이기 ( cat:0, dog:1)
# 데이터 한번에 csv에 밀어 넣으니까 데이터양이 너무 많아서 액셀이 못읽음 ㅜㅜ

# DataFrame 준비
df = list()

# cat 변환
# for n in range(5):
#     print('{}번째 시작'.format(n))
    
for i in range(1000):
    cat_img = np.array(Image.open(f'./data/cat_dog_small/train/cats/cat.{i}.jpg').resize((150,150)))
    dog_img = np.array(Image.open(f'./data/cat_dog_small/train/dogs/dog.{i}.jpg').resize((150,150)))

    for y in range(cat_img.shape[0]):
        for x in range(cat_img.shape[1]):
            cat_img[y,x] = int(np.mean(cat_img[y,x]))

        cat_gray_2d_img = cat_img[:,:,0]
        
    for y in range(dog_img.shape[0]):
        for x in range(dog_img.shape[1]):
            dog_img[y,x] = int(np.mean(dog_img[y,x]))
            
        dog_gray_2d_img = dog_img[:,:,0]

    df.append([0] + cat_gray_2d_img.ravel().tolist())
    df.append([1] + dog_gray_2d_img.ravel().tolist())   

    
df = pd.DataFrame(df)
df.to_csv(f'./data/cat_dog_data.csv', index=False)

```
